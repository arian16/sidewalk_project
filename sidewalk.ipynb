```python
# Create a df with open cases only
df2a = df.loc[df['case_status'] == 'Open', :]
```

```python
# df with pothole and sidewalk cases
cols_needed = ["Request for Pothole Repair", "Sidewalk Repair (Make Safe)", "Sidewalk Repair"]
df2b = df2a['type'].isin(cols_needed) # If the cols_need exits in df2a, df2b is boolean only
```

```python
df2c = df2a[df2b] # Only retain cols_need list
```

```python
df2c["type"].value_counts() 
```

```python
# Separate sidewalk and potholes
df_potholes = df2c.loc[df2c["type"] == "Request for Pothole Repair",:]
df_sidewalk = df2c.loc[df2c["type"] != "Request for Pothole Repair",:]
```

```python
fig1, ax1 = plt.subplots(figsize = (16,12))
ax1.scatter(df_potholes['longitude'], df_potholes['latitude'])
plt.show()
```

```python
fig1, ax1 = plt.subplots(figsize = (16,12))
ax1.scatter(df_sidewalk['longitude'], df_sidewalk['latitude'])
plt.show()
```

```python
# Adjust the center of the plot
pot_lat_center = np.float(df_potholes['latitude'].mean()) # Change numpy.float format to float
pot_lon_center = np.float(df_potholes['longitude'].mean())
# Scatter plot of potholes
gmap = gmplot.GoogleMapPlotter(pot_lat_center, pot_lon_center, 10)
%matplotlib inline
gmap.scatter(df_potholes['latitude'], df_potholes['longitude'], '# FF0000', size = 20, marker = False )
gmap.draw('potholes.html')
```

```python
# Adjust the center of the plot
side_lat_center = np.float(df_sidewalk['latitude'].mean()) # Change numpy.float format to float
side_lon_center = np.float(df_sidewalk['longitude'].mean())
# Scatter plot of potholes 
gmap = gmplot.GoogleMapPlotter(pot_lat_center, pot_lon_center, 10)
%matplotlib inline
gmap.scatter(df_sidewalk['latitude'], df_sidewalk['longitude'], '# FF0000', size = 20, marker = False )
gmap.draw('sidewalk_sidewalk.html')
```

```python
gmap2 = gmplot.GoogleMapPlotter(42.3244, -71.1147, 16)
lat = [-71.0867, -71.0867]
lon = [42.3435, 42.3423]
# Function to calculate average of a list
def average_list(list_input): 
    return sum(list_input) / len(list_input) 
ave_lat = average_list(lat)
ave_lon = average_list(lon)
gmap2 = gmplot.GoogleMapPlotter(ave_lat, ave_lon, 16)
gmap2.plot( lat, lon, 'blue', edge_width = 1 )
gmap2.draw('pilot2.html')
```
